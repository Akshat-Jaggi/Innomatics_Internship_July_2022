```python
import pandas as pd
import numpy as np
```

```python
col = pd.read_csv("data_dictionary.csv")
col.head()
```

```python
data = pd.read_csv('open_pubs.csv')
```

```python
data.head()
```

```python
new_header = col["Field"] 
data.columns = new_header 
data.head(10)
```

```python
data.loc[0] = ['22','Anchor Inn', 'Upper Street, Stratford St Mary, COLCHESTER', 'CO7 6LW', '604749', '234404', '51.970379', '0.97934','Babergh' ]
```

```python
data.head(10)
```

```python
data.isna().sum()
```

```python
data.drop(data.index[data['latitude'] == '\\N' ], inplace=True)
```

```python
data.head(10)
```

```python
data.reset_index(drop=True,inplace = True)
```

```python
data.head()
```

```python
data.dtypes
```

```python
data[['latitude', 'longitude']] = data[['latitude', 'longitude']].apply(pd.to_numeric)
```

```python
data[["name","address","postcode","local_authority"]] = data[["name","address","postcode","local_authority"]].astype("string")#string
data[["fsa_id","easting","northing"]] = data[["fsa_id","easting","northing"]].astype("int") #to interger
```

```python
data.dtypes
```

```python
data.to_csv('modified_open_pubs.csv',index=False)
```
